```r
media = 2
desvpad = 0.6
x = 1.5
```

### Draw the normal distribution

```r
range = seq(
            media - 4*desvpad,
            media + 4*desvpad, 
            0.01
          )
```

```r
y = dnorm(
          range, 
          media, 
          desvpad
        )
```

```r
plot(
      range,
      y,
      main =,
      type = 'l',
      ylim = c(0,max(y) + 0.01),
      axes = T
    )
```

```r
plot(
      range,
      y,
      main =,
      type = 'l',
      ylim = c(0,max(y) + 0.01),
      axes = F
    )
axis(
      1, 
      at = seq(
                media - 4*desvpad, 
                media + 4*desvpad,
                desvpad
              )
    )
```

### Add area to the left of x

```r
cord.a = c(
            min(range),
            seq(min(range), x, 0.01),
            x                
          )
```

```r
cord.b = c(
            min(range),
            dnorm(
                  seq(min(range), x, 0.01),
                  media,
                  desvpad
                ),
            0
          )
```

```r
plot(
      range,
      y,
      main =,
      type = 'l',
      ylim = c(0,max(y) + 0.01),
      axes = F
    )

axis(
      1, 
      at = seq(
                media - 4*desvpad, 
                media + 4*desvpad, 
                desvpad
              )
    )

polygon(cord.a, cord.b, col = "blue")
```

### Add area to the right side of x

```r
cord.c = c(
            x, 
            seq(x, 
                max(range),
                0.01
              ),
            0
          )
```

```r
cord.d = c(
            0,
            dnorm(
                  seq(
                      x,
                      max(range),
                      0.01
                    ),
                  media,
                  desvpad
                ),
              0
            )
```

```r
plot(
      range,
      y,
      main =,
      type = 'l',
      ylim = c(0,max(y) + 0.01),
      axes = F
    )

axis(
      1, 
      at = seq(
                media - 4*desvpad, 
                media + 4*desvpad, 
                desvpad
              )
    )

polygon(cord.c, cord.d, col = grey(0.90))
```

### P(X < x) Area to the left of x

```r
round(pnorm(x, media, desvpad)*100,2)

```

### P(X > x) Area to the right of x

```r
round(100 - pnorm(x, media, desvpad)*100,2)
```

## P(x =< X <= a) Area between "x" and "b"

```r
a = 3
round((pnorm(a, media, desvpad) - pnorm(x, media, desvpad))*100,2)
```

## Normal Distribution, Z Scores, and Normal Probabilities

```r
# P(X <= 70)
round(
       pnorm(
              q=70, 
              mean=75, 
              sd=5, 
              lower.tail = T
            )*100,         
       2
      )
```

```r
# P(X <= 70)
round(
       pnorm(
              q=70, 
              mean=75, 
              sd=5, 
              #lower.tail = T
            )*100,         
       2
      )

#lower.tail by default
```

```r
round(
       pnorm(
              85, 
              75, 
              5, 
              lower.tail = F
            )*100,         
       2
      )
```

## P(Z >= 1)

```r
round(
       pnorm(
              1, 
              0, #reduced Gaussian distribution
              1, #reduced Gaussian distribution
              lower.tail = F
            )*100,         
       2
      )
```

## Find Quantiles

```r
round(
       qnorm(
              .25, #Q1
              75, 
              5, 
              lower.tail = T
            ),         
       2
      )
```

```r
x = seq(
        from = 55, 
        to = 95, 
        by = .25
        )
```

```r
print(x)
```

```r
dens = dnorm(
              x, 
              mean(x), 
              sd(x)
            )
```

```r
plot(
      x,
      dens,
      type = 'l',
      main = "X-Normal",
      xlab = "x",
      ylab = "Probability Density",
      las = 1
    )

abline(v = mean(x))
```

```python
import statsmodels.api as sm
```

```python
depVar = data['quality']
indepVar = data.drop(labels = 'quality', axis = 1)
indepVar
```

```python
depVar = data['quality']
indepVar = data.drop(labels = 'quality', axis = 1)

indepVar = sm.add_constant(indepVar) # must include in the model
model = sm.OLS(depVar, indepVar).fit()
print(model.summary())
```

```python
model.pvalues
```

```python
model.pvalues<.05
```

```python
model.pvalues[model.pvalues<.05].keys()
```

```python
list(model.pvalues[model.pvalues<.05].keys())
```

```python
sigCols = list(model.pvalues[model.pvalues<.05].keys())
[print(i) for i in sigCols];
```

```python
binthresh = np.mean(data['quality'])
binthresh
```

```python
data['quality'] > binthresh
```

```python
data['binquality'] = data['quality'] > binthresh
data
```

```python
Xcols = []

for key in data.keys():
  if key not in ['quality', 'binquality']:
    Xcols.append(key)

Xcols
```

```python
model = sm.Logit(data['binquality'], data[Xcols])
results = model.fit(method = 'newton')
results.summary()
```

```python
list(results.pvalues[results.pvalues<.05].keys())
```

```python
sigColsL = list(results.pvalues[results.pvalues<.05].keys())
```

```python
n = 500

x = np.cumsum(np.random.randn(n))
y = stats.rankdata(x)

y
```

```python
n = 500

x = np.cumsum(np.random.randn(n))
y = stats.rankdata(x)

np.min(y)
```

```python
n = 500

x = np.cumsum(np.random.randn(n))
y = (stats.rankdata(x)/(n+1) - .5)*2 # normalization (values from -1 to 1)

print(np.min(y), np.max(y))
```

```python
# Algorthm to transform a non-normal distribution to a Gaussian distribution

n = 500

x = np.cumsum(np.random.randn(n))
y = (stats.rankdata(x)/(n+1) - .5)*2
y = np.arctanh(y)
```

```python
fig,ax = plt.subplots(2,2,figsize=(8,8))

ax
```
